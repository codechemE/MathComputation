```python
import matplotlib.pyplot as plt
import numpy as np


def func(x,y):
    return y+2*x-x**2

x0 = 0
y0 = 1

x, h = np.linspace(x0, 1.5, 50, retstep=True)
# initialize the sol
y = np.zeros(x.shape)
y[0] = y0

for i in range(0, len(x) - 1):
    y[i+1] = y[i] + func(x[i],y[i])*h

plt.plot(x, y, label = 'Euler')
plt.plot(x, x**2+np.exp(x), 'r', label = 'Analytical')
```

```python
import matplotlib.pyplot as plt
import numpy as np


def func(x,y):
    return y+2*x-x**2

x0 = 0
y0 = 1

x, h = np.linspace(x0, 1.5, 5, retstep=True)
y = np.zeros(x.shape)
y[0] = 0

for i in range(0, len(x) - 1):
    k1 = h*func(x[i],y[i])
    k2 = h*func(x[i]+h/2,y[i]+k1/2)
    k3 = h*func(x[i]+h/2,y[i]+k2/2)
    k4 = h*func(x[i+1]+h,y[i]+h*k3)
    y[i+1] = y[i] + h / 6 * (k1 + 2*k2 + 2*k3 + k4)
    
plt.plot(x, y, label="Runge-Kutta Method")
```
