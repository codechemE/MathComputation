```python
import numpy as np

A = np.array([[2, 7, -1, 3, 1],
              [2, 3, 4, 1, 7],
              [6, 2, -3, 2, -1],
              [2, 1, 2, -1, 2],
              [3, 4, 1, -2, 1]], float)

b = np.array([5, 7, 2, 3, 4], float)

n = len(b)

x = np.zeros(n, float)

for k in range(0, n - 1):
    if A[k, k] == 0:
        for j in range(n):
            A[k, j], A[k + 1, j] = A[k + 1, j], A[k, j]
        b[k], b[k + 1] = b[k + 1], b[k]
    for i in range(k + 1, n):
        lam = A[i, k] / A[k, k]
        b[i] = b[i] - lam * b[k]

        for j in range(k, n):
            A[i, j] = A[i, j] - lam * A[k, j]

x[n - 1] = b[n - 1] / A[n - 1, n - 1]
for i in range(n - 2, -1, -1):
    terms = 0
    for j in range(i + 1, n):
        terms += A[i, j] * x[j]
    x[i] = (b[i] - terms) / A[i, i]
print(x)

```
