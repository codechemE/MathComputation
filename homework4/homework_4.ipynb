```python
import numpy as np
import matplotlib.pyplot as plt

time = np.array([0, 5, 10, 15, 20])
concentration = np.array([.1, .0141,.0078, .0053, .004])
ln_concentration = np.log(concentration)
a_second_order = 1/concentration

fig1, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(time, ln_concentration)
ax2.plot(time, concentration)

fig2,(ax3) = plt.subplots(1, 1)
ax3.plot(time, a_second_order)

```
